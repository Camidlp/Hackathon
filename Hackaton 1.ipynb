### 6) Modelo de Regresion Lineal

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(htrain.drop(['SalePrice'],axis=1),htrain['SalePrice'],test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression(fit_intercept = True)
lr.fit(x_train, y_train)
```

```python
print(lr.coef_)
print(lr.intercept_)
print(lr.rank_)
```

```python
from sklearn.metrics import mean_squared_log_error
y_train_pred = lr.predict(x_train)
y_test_pred = lr.predict(x_test)
print('Error en datos de train:', mean_squared_log_error(y_train, y_train_pred, squared=False))
print('Error en datos de test:', mean_squared_log_error(y_test, y_test_pred, squared = False))
```

```python
from sklearn.pipeline import Pipeline
pipe = Pipeline([('LR', LinearRegression())])
pipe.fit(x_train, y_train)
Pipeline(steps=[('LR', LinearRegression())])
pipe.score(x_test, y_test)
```
